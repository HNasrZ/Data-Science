## A function which handles the multiple choice questions
- Replace not selected choices with 0 and selected ones with 1
- It also removed the "text" column(s)
- It removes the "None" choice column, if existed. This means that if a respondant has not answered to the question, it is assumed that he/she uses "none" of the choices.

```python
def replace_nan_with_0(df,question_number,parts,has_other_text=True,drop_parts=[]):

    list_question_parts = []
    for part in range(1,parts+1):
        if part not in drop_parts:
            list_question_parts.append('Q' + str(question_number) + '_Part_' + str(part))
    
    qs_df = df[list_question_parts]

    for question in list_question_parts:
        len_null= len(qs_df[qs_df[question].isnull()])
        len_not_null=len(qs_df[question])-len_null
        
        if  len_not_null < 100:
            qs_df=qs_df.drop(question,axis=1)
    
    #set the columns of the data frame as the name of the part of question 
    columns=qs_df.mode().values[0]

    colList=[]
    for i in columns:
        colList.append('Q' + str(question_number) + '_'+ i)
        
    qs_df.columns = colList
    
    #fill the NaN cells with zero, which means that the respondant did not check this part of the question
    qs_df_binary = qs_df.fillna(0).replace('[^\\d]',1, regex=True) 
    
    for i in colList:
        df[i]=qs_df_binary[i]
    
    for i in list_question_parts:
        df=df.drop(i,axis=1)
    
    if has_other_text:
        df=df.drop('Q' + str(question_number) + '_OTHER_TEXT',axis=1)
        
    for i in drop_parts:
        df=df.drop('Q' + str(question_number) + '_Part_' + str(i),axis=1)

    return df
```

### Define a function which performs 10-fold croos validation

```python
#set up cross validation
def run_kfold(model):
    
    X = Salaries_X
    Y = Salaries_Y
    
    kf = KFold(n_splits=10) #n_splits previously n_folds
    
    outcomes = []
    fold = 0
    
    for train_index, test_index in kf.split(X):
        fold += 1
        X_train, X_test = X.values[train_index], X.values[test_index]
        Y_train, Y_test = Y.values[train_index], Y.values[test_index]
        
        model.fit(X_train, Y_train)
        predictions = model.predict(X_test)
        
        accuracy = r2_score(Y_test, predictions) # can try mean absolute error instead
        outcomes.append(accuracy)
        print("Fold {0} accuracy: {1}".format(fold, accuracy))   
        
#         print(list(X_test))
    mean_outcome = np.mean(outcomes)
    std_outcome=np.std(outcomes)
    print("Mean r2: {0}".format(mean_outcome)) 
    print("Standard Deviation: {0}".format(std_outcome))
    
    return (mean_outcome,std_outcome)
```

The main use of linear regression is to find out the relationship between variables andd with the target variable. The linear regression technique finds out a linear relationship between x (input) and y(output).
<br>The general formula of linear regression is:
<br> $y$ = $\theta_0$ + $\theta_1$*$x_1$ + $\theta_2$*$x_2$ + ... + $\theta_n$$x_n$+$\epsilon$
<br> The $\Theta$ = {$\theta_i$} are model parameters, $X$={$x_i$} are features, and the $\epsilon$ is the model error.

### Pros

- It can identify outliers and anomolies
### Cons

## 4.1. Explain how feature engineering is a useful tool in machine learning.

Data is the most important part of any machine learning algorithm. However, the raw data needs to be processed and refined into features - predictor variables - to be useful for developing and training models. Without features it is not possible to train the accurate model regardless of the complexity of the machine learning algorithm. The process of extracting features from a raw dataset is called **feature engineering**.

**The Feature Engineering Process**:

Feature engineering means building features for the purpose of model development. These features are then passed to modeling where they will be used for training a machine learning algorithm.

Conventianlly, feature engineering is done manually, using domain knowledge. However, this manual process is error-prone, tedious, must be started from scratch for each dataset. Regarding that, there are some automated feature engineering algorithms which can automatically build hundreds of relevant features from a relational dataset. The **better the feature engineering** is done, the scientists are more enabled to develop better models with **better prediction**.

## 4.2. Do the feature selection - Recursive Feature Elimination

### Description

Given an external estimator that assigns weights to features (e.g., the coefficients of a linear model), the goal of recursive feature elimination (RFE) is to select features by recursively considering smaller and smaller sets of features. First, the estimator is trained on the initial set of features and the importance of each feature is obtained either through a coefficient attribute or through a feature importances attribute. Then, the least important features are pruned from current set of features. That procedure is recursively repeated on the pruned set until the desired number of features to select is eventually reached.

## 4.3. Lasso Regularization

### Description

Lasso (Least Absolute Shrinkage and Selection Operator) regression is a powerful technique generally used for creating parsimonious models in presence of a ‘large’ number of features. Large number features means:
- Large enough to enhance the tendency of a model to overfit. This is a problem that we have here in our problem.
- Large enough to cause computational challenges. This problem does not apply to our problem here.

Lasso Regression performs L1 regularization, i.e. adds penalty equivalent to absolute value of the magnitude of coefficients

Minimization objective = LS Obj + $\alpha$ * (sum of absolute value of coefficients)
Note that here ‘LS Obj’ refers to ‘least squares objective’, i.e. the linear regression objective without regularization.

Adapted from https://www.analyticsvidhya.com/blog/2016/01/complete-tutorial-ridge-lasso-regression-python/



```python
from sklearn import linear_model
reg_lasso = linear_model.Lasso(alpha = 0.5,max_iter=10000) # optimal parameters are entered

reg_lasso.fit(Salaries_X, Salaries_Y)
reg_lasso.score(Salaries_X, Salaries_Y)
```

### 4.3.2. Hyper Parameter tuning (Manual Grid Search) for Lasso Regression without Cross validation 

```python
# Compute train and test errors
alphas=[0.5,1,3,5,10,20,50,100,200,500,1000,5000,10000,20000,50000,100000]
reg_lasso1 = linear_model.Lasso()
train_errors = list()
test_errors = list()
for alpha in alphas:
    reg_lasso1.set_params(alpha=alpha)
    reg_lasso1.fit(Salaries_X, Salaries_Y)
    train_errors.append(reg_lasso1.score(Salaries_X, Salaries_Y))
    test_errors.append(reg_lasso1.score(Salaries_X_Test, Salaries_Y_Test))

i_alpha_optim = np.argmax(test_errors)
alpha_optim = alphas[i_alpha_optim]
print("Optimal regularization parameter : %s" % alpha_optim)

# Estimate the coef_ on full data with optimal regularization parameter
reg_lasso1.set_params(alpha=alpha_optim)
lasso_coef_ = reg_lasso1.fit(Salaries_X, Salaries_Y).coef_
scoreOptim = reg_lasso1.score(Salaries_X_Test, Salaries_Y_Test)
print("Optimal score : %s" % scoreOptim)
```

```python
plt.subplots(1, 2, sharey='row', figsize=(12, 12))

plt.subplot(211)
plt.semilogx(alphas, train_errors, label='Train')
plt.semilogx(alphas, test_errors, label='Test')
plt.vlines(alpha_optim, plt.ylim()[0], np.max(test_errors), color='k',
           linewidth=3, label='Optimum on test')
plt.legend(loc='lower left')
plt.ylim([0, 1.3])
plt.xlabel('Regularization parameter',fontsize=18)
plt.ylabel('Performance (Score)',fontsize=18)

# Show estimated coef_ vs true coef
plt.subplot(212)

plt.plot(lasso_coef_, label='Estimated coef')
plt.xlabel('Feature index',fontsize=18)
plt.ylabel('Estimated Coefficinets',fontsize=18)

plt.legend()
plt.subplots_adjust(0.09, 0.04, 0.94, 0.94, 0.26, 0.26)
plt.show()
```

### Discussion
According to the first figure, as the $\alpha$ paramter increases, the training score decreases. However, the test score increases to some point and then decreases. Therefore, as the above analysis suggest, the optimum $\alpha$ parameter which results in highest accuracy on test data set is 200. Therefore, the optimal accuracy score is 56.1 % which is slightly better than the regular linear regression.

## 3.3. Decision Tree Regression
Decision tree builds regression models in the form of a tree structures. It breaks down a dataset into smaller subsets while at the same time an associated decision tree is incrementally developed. The final result is a tree with decision nodes and leaf nodes. A decision node has two or more branches, each representing values for the attribute tested. Leaf node represents a decision on the numerical target. The topmost decision node in a tree which corresponds to the best predictor called root node. Decision trees can handle both categorical and numerical data. 

#### Pros
- Too much insightful information


#### Cons
- Complexity
- Computation cost
- Inadequacy in applying regression and predicting continuous values
- Possibility of spurious relationships


### Description
